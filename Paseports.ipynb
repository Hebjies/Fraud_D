```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
Paseport = cv.imread("Colombia1.png") # Reads the photo saved in specific location
# Para dejar todo impreso en el doc, se usa plt en vez de cv.
#cv.imshow("Pasaporte", Paseport)
plt.imshow(Paseport)
plt.title("Ejemplo Pasaporte #1")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```
